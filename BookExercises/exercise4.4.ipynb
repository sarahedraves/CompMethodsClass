```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpoints=100
```

```python
xs=np.linspace(0,1,numpoints)
```

```python
deltax=xs[1]-xs[0]
```

```python
condlist=[xs<1/3, xs<2/3, xs<1]
choicelist=[0,1,0]
a0=np.select(condlist,choicelist)
```

```python
u=1
C=0.7
deltat=C*deltax/u
```

```python
period=1/u
periods=1
totaltime=period*periods
numsteps=int(np.ceil(totaltime/deltat))
```

```python
an=a0
for n in range(numsteps):
    apadded=np.pad(an,1,mode='wrap')
    an1=np.zeros(numpoints)
    for i in range(numpoints):
        an1[i]=-u*deltat*(apadded[i+2]-apadded[i])/deltax + apadded[i+1] #indices are shifted by 1 for padded array
    an=an1
```

```python
np.round(an,2)
```

now do gaussian initial conditions using formula e^-x^2

```python
gpoints=np.exp(-((xs-.5)**2)/0.2)
```

```python
plt.scatter(xs,gpoints)
```

```python
an=gpoints
for n in range(numsteps):
    apadded=np.pad(an,1,mode='wrap')
    an1=np.zeros(numpoints)
    for i in range(numpoints):
        an1[i]=-u*deltat*(apadded[i+2]-apadded[i])/deltax + apadded[i+1]
    an=an1
```
